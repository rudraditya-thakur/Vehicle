```python
from IPython.display import display
import ipywidgets as widgets

def play_mp4_video(video_path):
    video_widget = widgets.Video(
        value=open(video_path, 'rb').read(),
        format='mp4',
        width=1280,
        height=720
    )
    display(video_widget)
```

```python
play_mp4_video('/kaggle/input/hwid12-highway-incidents-detection-dataset/Video-Accident-Dataset/pedestrian_hit/pedestrian_hit_10.mp4')
```
